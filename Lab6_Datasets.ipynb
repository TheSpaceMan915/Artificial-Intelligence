# **Selecting a target variable**
A target variable is a column of a dataset about which we would like to gain a deeper understanding

```python
df_dropped = df.drop("rating", 1)
s_target = df["rating"]
df_dropped.head()
```

# **Changing values that are not appropriate to NaNs**
Filtering release years that are greater than 2023 and less than 1950

```python
# finding not appropriate values
mask_booleans = df_dropped["release_year"].lt(2023) & df_dropped["release_year"].ge(1970)
df_dropped["release_year"] = df_dropped["release_year"].where(mask_booleans, np.nan)

print("Amount of NaNs:", df_dropped["release_year"].isnull().sum())
```

# **Processing NaN values**

```python
df_dropped["duration"] = df_dropped["duration"].fillna("Undefined")
df_dropped["director"] = df_dropped["director"].fillna("Undefined")
df_dropped["cast"] = df_dropped["cast"].fillna("Undefined")
df_dropped["country"] = df_dropped["country"].fillna("Undefined")
df_dropped["release_year"] = df_dropped["release_year"].fillna(2021)
df_dropped.head()
```

# **Looking for implicit duplicates**

```python
def clean_duration(elem, list1, list2):
  if elem.lower().rstrip() in list1: 
    return "1 season"
  elif elem.lower().rstrip() in list2:
    return "2 seasons"
  else:
    return "Other"


list_one_season = ["1 season", "1 Season", "one season", "One season"]
list_two_seasons = ["2 seasons", "2 Seasons", "two seasons", "Two seasons"] 


df_dropped["duration"] = df_dropped["duration"].apply(
    lambda x : clean_duration(x, list_one_season, list_two_seasons))
df_dropped.head()
```

# **Data dependencies**
If we'd like to know what dependencies there are in our data, we ought to create a correlation map

```python
var_corr = round(df_dropped.corr(),2)
print(var_corr)

mask = np.zeros_like(var_corr)
heatmap = sns.heatmap(var_corr
                , mask = mask
                , square = True
                , annot = True
                , annot_kws={'size': 10.5, 'weight' : 'bold'}
                , cmap=plt.get_cmap("RdPu")
                , linewidths=.1)
heatmap.figure.set_size_inches(8,9)

plt.title('Correlation map', fontsize=14)
plt.show()
```

# **Standart data normalisation**
Normalization is a rescaling of the data from the original range so that all values are within the new range of 0 and 1.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_dropped[["release_year"]] = scaler.fit_transform(df_dropped[["release_year"]])
df_dropped.head()
```

```python
# getting not normalised data back to the column
df_dropped["release_year"] = df["release_year"]
df_dropped["release_year"] = df_dropped["release_year"].fillna(2021)
```

# **Min-max data normalisation**

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df_dropped[["release_year"]] = scaler.fit_transform(df_dropped[["release_year"]])
df_dropped.head()
```

# **Dividing data for learning and testing**

```python
from sklearn.model_selection import train_test_split

# getting several lists of data from the dataframe
# s_target is the result series with the outputs (the thing we want to learn about)
list_df_train, list_df_test, list_s_train, list_s_test = train_test_split(df_dropped, s_target, test_size=0.2, random_state=0)

print("The shapes of the lists: ")
print("list_df_train:", list_df_train.shape)
print("list_df_test:", list_df_test.shape)
print("list_s_train:", list_s_train.shape)
print("list_s_test:", list_s_test.shape)
```
