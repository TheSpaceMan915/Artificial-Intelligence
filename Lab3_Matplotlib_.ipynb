# **1. Измените положение легенды в Matplotlib**

# **2. Как отобразить значение каждого столбца на гистограмме с помощью Matplotlib?**

```python
# creating data
x = np.random.randn(1000)

# creating an empty canvas and axes
fig, axes = plt.subplots(1, figsize = (10,6))

# creating a histogram
counts, edges, bars = plt.hist(x)
axes.set_xlim((min(x), max(x)))
axes.set_title("Histogram")

# creating x ticks
xticks = [(edges[index+1] + value)/2 for index, value in enumerate(edges[:-1])]

# showing values on the bars
for index, value in enumerate(counts): 
  plt.text(xticks[index], value, int(value), horizontalalignment="center")
```

# **3. Как аннотировать столбцы в Barplot с помощью Matplotlib на Python?**

```python
# creating data
x = np.array([i for i in range(10)])

# creating an empty canvas and axes
fig, axes = plt.subplots(1, figsize = (10,12))

# creating a bar graph
axes.bar(x, x + 2, align = "center", width = 0.5, alpha = 0.4)
axes.set_title("Bar")

# creating patches
arr_rectangles = axes.patches

# creating a few labels
arr_labels = [f"Label{i}" for i in range(len(x))]

# putting the labels on the bars
for rect, label in zip(arr_rectangles, arr_labels):
  height = rect.get_height()
  axes.text(rect.get_x() + rect.get_width() / 2, height + 0.2, label, ha = "center", va = "bottom")
```

# **4. Как аннотировать столбцы в сгруппированном столбчатом графике на Python?**

```python
# creating data
arr_number_cats = [1,3,1,2,2]
arr_number_dogs = [2,3,2,1,1]

# initialising the locations of the labels and the width between the bars
x = np.arange(len(arr_labels))
width = 0.3

# creating an empty canvas and axes
fig, axes = plt.subplots(1, figsize = (10,12))

# changing the font and its size
plt.rcParams.update({"font.size": 16, "font.family": "serif"})

# creating a bar graph
rectangles1 = axes.bar(x-width/2, arr_number_cats, width, label = "Cats")
rectangles2 = axes.bar(x+width/2, arr_number_dogs, width, label = "Dogs")

# adding titles and labels
axes.set_ylabel("Number of animals", fontsize = "medium")
axes.set_title("Family pets")
axes.legend()

# adding bar labels
for rect, label in zip(rectangles1, arr_number_cats):
  height = rect.get_height()
  axes.text(rect.get_x() + rect.get_width() / 2, height + 0.05, label, ha = "center", va = "bottom")

for rect, label in zip(rectangles2, arr_number_dogs):
  height = rect.get_height()
  axes.text(rect.get_x() + rect.get_width() / 2, height + 0.05, label, ha = "center", va = "bottom")
```

```python
# creating labels
arr_genres = ["Education", "Comics", "History", "Fantasy"]

# creating sizes
arr_sizes = [25, 10, 10, 55]
tuple_explodes = (0.1, 0, 0, 0)

# creating an empty canvas and axes
fig, axes = plt.subplots(1, figsize = (6,8))

# changing the font and its size
plt.rcParams.update({"font.size": 18, "font.family": "STIXGeneral", "mathtext.fontset": "stix"})

# creating a pie chart
axes.pie(arr_sizes, explode = tuple_explodes, labels = arr_genres, autopct = "%1.1f%%" ,shadow = True)
axes.set_title("Pie chart")
axes.axis("equal")
```

# **2. Как установить границу для клиньев в круговой диаграмме Matplotlib?**

```python
# creating labels
arr_genres = ["Education", "Comics", "History", "Fantasy"]

# creating sizes
arr_sizes = [25, 10, 10, 55]
tuple_explodes = (0.1, 0, 0, 0)

# creating an empty canvas and axes
fig, axes = plt.subplots(1, figsize = (6,8))

# changing the font and its size
plt.rcParams.update({"font.size": 18, "font.family": "STIXGeneral", "mathtext.fontset": "stix"})

# creating a pie chart and setting up the borders using wedgeprops attribute
axes.pie(arr_sizes, explode = tuple_explodes, labels = arr_genres, autopct = "%1.1f%%" ,shadow = True, startangle = 90, 
         wedgeprops = {"linestyle": "-", "linewidth": 3, "edgecolor": "black"})
axes.set_title("Pie chart")
axes.axis("equal")
```
